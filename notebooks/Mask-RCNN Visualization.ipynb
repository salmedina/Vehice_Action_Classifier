```python
import pycocotools.mask as maskUtils
import json
import os
from PIL import Image
from matplotlib.pyplot import imshow
from IPython.display import display
%matplotlib inline
```

```python
data_path = '/home/zal/Data/VIRAT/Output/objdetector/train/'
sample_filename = 'VIRAT_S_000000_F_00003008.json'
```

```python
filename = os.path.join(data_path, sample_filename)
data = json.load(open(filename))
n = len(data)
print(''n)
for i in range(n):
    s = data[i]
    if s['cat_name'] == 'car':
        print('Score: %f'%(s['score']))
        val = maskUtils.decode(s['segmentation']) * 255
        img = Image.fromarray(val)
        display(img)
```
