```python
def get_id_class_dict(train_ids_list, test_ids_list):
    id_class_dict = {}
    for c, video_ids_list in enumerate(train_ids_list):
        for video_id in video_ids_list:
            id_class_dict[video_id] = c
    
    for c, video_ids_list in enumerate(test_ids_list):
        for video_id in video_ids_list:
            id_class_dict[video_id] = c
            
    return id_class_dict

def get_dir_video_id(dir_name):
    return dir_name[8:14]

def copy_first_frames(source_dir, target_dir, n):
    name_list = os.listdir(source_dir)
    for frame_num in range(1,n+1):
        source_name ='%d.jpg'%frame_num
        if source_name in name_list:
            shutil.copy(osp.join(source_dir, source_name), target_dir)
```

```python
source_dir = '/home/zal/Data/VIRAT/Frames/imgs'
target_dir = '/home/zal/Data/VIRAT/Frames/first_frames'

class_dir = get_id_class_dict(train_ids_list, test_ids_list)
clip_annotations = []
for clip_dir in os.listdir(source_dir):
    clip_class = class_dir[get_dir_video_id(clip_dir)]
    clip_annotations.append((clip_dir, clip_class))
    os.makedirs(osp.join(target_dir, clip_dir))
    copy_first_frames(osp.join(source_dir, clip_dir), osp.join(target_dir, clip_dir), 30)

with open(osp.join(target_dir, 'annotations.csv'),'w') as out:
    csv_out = csv.writer(out)
    csv_out.writerow(['dir_name','scene_class'])
    for row in clip_annotations:
        csv_out.writerow(row)
        
clip_annotations
```
