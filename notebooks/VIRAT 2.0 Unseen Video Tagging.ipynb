```python
import cv2
import os
import numpy as np
import os.path as osp
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt
import pickle
```

```python
virat2_frames_root = '/mnt/Alfheim/Data/VIRAT2/VIRAT-V2-drop1/Frames'
```

```python
def get_video_names(frames_path):
    return os.listdir(frames_path)

def get_first_frame_paths(frames_path):
    return glob(osp.join(frames_path, '*','1.jpg'))

def load_first_frames(frames_path, size):
    frames_paths = get_first_frame_paths(frames_path)
    frames = map(Image.open, frames_paths)
    return frames_paths, [np.array(f.resize(size)) for f in frames]
```

```python
frames_paths, frames = load_first_frames(virat2_frames_root, (192,108))
```

```python
total_frames = len(frames)
mosaic_h = 10
mosaic_w = 15

img_strips = []
for start_idx in np.arange(0, total_frames, mosaic_w):
    img_strips.append(np.hstack(frames[start_idx:start_idx+mosaic_w]))

mosaic = Image.fromarray(np.vstack(img_strips))
```
