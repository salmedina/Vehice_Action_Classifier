```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = np.matrix([[-0.00025796104,0.00093565491,-0.49343613],
            [0.00057183404,0.00082778391,-0.86963374],
            [1.5464549e-006,4.0565615e-005,0.016012838]])
```

```python
img = cv2.imread('/home/zal/Data/VIRAT/V2/Homographies/VIRAT_S_0000.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
warped_img = cv2.warpPerspective(img, H, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(warped_img)
plt.show()
```
